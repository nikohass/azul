```python
def generate_combinations():
    from itertools import combinations_with_replacement
    combinations = []
    for comb in combinations_with_replacement(range(5), 4):
        distribution = [0] * 5
        for i in comb:
            distribution[i] += 1
        combinations.append(distribution)
    return combinations

INDEX_TO_FACTORY = sorted([tuple(f) for f in generate_combinations()])
FACTORY_TO_INDEX = {f: i for i, f in enumerate(INDEX_TO_FACTORY)}
```

```python
import torch
import torch.nn as nn

def get_input_vector(game_state):
    factory_vector_size = (azul.NUM_FACTORIES - 1) * len(INDEX_TO_FACTORY) * 5
    pattern_line_vector_size = (1 + 2 + 3 + 4 + 5) * azul.NUM_TILE_COLORS

    player_wall_vector_size = 2 ** 5 * 5 * 3
    player_vector_size = pattern_line_vector_size + player_wall_vector_size

    total_input_size = factory_vector_size + player_vector_size * azul.NUM_PLAYERS
    
    vector = np.zeros(total_input_size, dtype=float)

    factories = np.array(game_state.factories)
    count = {}
    for factory in factories[:azul.NUM_FACTORIES - 1]:
        index = FACTORY_TO_INDEX.get(tuple(factory))
        if index is None:
            continue
        if index not in count:
            count[index] = 0
        count[index] += 1

    for factory_index, number in count.items():
        index = factory_index + (azul.NUM_FACTORIES - 1) * factory_index
        vector[index * 5 + number] = 1.0

    offset = factory_vector_size

    for player in range(0, azul.NUM_PLAYERS):
        cumulative_sizes = np.cumsum([i * azul.NUM_TILE_COLORS for i in range(1, 6)])
        for pattern_line_index, color in enumerate(pattern_line_colors):
            if color is None: 
                continue
            fill_status = pattern_lines[pattern_line_index]
            base_index = cumulative_sizes[pattern_line_index - 1] if pattern_line_index > 0 else 0
            color_offset = int(color) * (pattern_line_index + 1) 
            index = base_index + fill_status + color_offset
            vector[offset + index] = 1.0

        offset += pattern_line_vector_size
        wall = game_state.walls[player]
        for row in range(5):
            index = wall >> (6 * row) & 0b11111
            # player_wall_vector[index + 32 * row] = 1
            index = offset + index + 32 * row
            vector[index] = 1.0

        for col in range(5):
            index = 0
            for row in range(5):
                # Extract the bit at the current column and row
                bit = (wall >> (6 * row + col)) & 1
                # Shift the bit to its position in the 5-bit integer
                index |= bit << row
            # Set the bit in the vector
            # player_wall_vector[index + 32 * col + 2 ** 5 * 5] = 1  # Offset by 5 * 32 to avoid overlap with rows
            index = offset + index + 32 * col + 2 ** 5 * 5
            vector[index] = 1.0

        for diag_index, mask in enumerate(azul.WALL_COLOR_MASKS):
            index = 0
            masked = wall & mask
            for row in range(5):
                bit = (masked >> (6 * row)) & 0b11111
                index |= bit
            # player_wall_vector[index + 32 * diag_index + 2 ** 5 * 10] = 1  # Offset by 10 * 32 to avoid overlap with rows and columns
            index = offset + index + 32 * diag_index + 2 ** 5 * 10
            vector[index] = 1.0

        offset += player_wall_vector_size
    
    return vector

class AzulNnue(nn.Module):
    def __init__(self):
        super(AzulNnue, self).__init__()

vec = get_input_vector(game_state)
```

```python
import numpy as np

CENTER_FACTORY_ENCODING_SIZE = NUM_TILE_COLORS * 10
FACTORY_ENCODING_SIZE = (NUM_FACTORIES - 1) * len(INDEX_TO_FACTORY) * 5

PATTERN_LINE_ENCODING_SIZE = (1 + 2 + 3 + 4 + 5) * NUM_TILE_COLORS
WALL_ENCODING_SIZE = 2 ** 5 * 5 * 3

TOTAL_ENCODING_SIZE = FACTORY_ENCODING_SIZE + CENTER_FACTORY_ENCODING_SIZE + \
    (WALL_ENCODING_SIZE + PATTERN_LINE_ENCODING_SIZE) * NUM_PLAYERS

def get_factory_indices(factories):
    count = {}
    for factory in factories[:NUM_FACTORIES - 1]:
        index = FACTORY_TO_INDEX.get(tuple(factory))
        if index is None:
            continue
        if index not in count:
            count[index] = 0
        count[index] += 1

    indices = []
    for index, number in count.items():
        indices.append(index * 5 + number)

    return indices

def factories_from_indices(indices):
    factories = np.zeros((NUM_FACTORIES, 5), dtype=int)
    current_index = 0
    for index in indices:
        factory_tuple = INDEX_TO_FACTORY[index // 5]
        factories[current_index] = factory_tuple
        current_index += 1
    return factories

def pattern_lines_to_indices(pattern_line_occupancy, player_pattern_line_colors):
    assert len(pattern_line_occupancy) == 5
    assert len(player_pattern_line_colors) == 5

    cumulative_sizes = np.cumsum([i * NUM_TILE_COLORS for i in range(1, 6)])

    indices = []
    for pattern_line_index, color in enumerate(player_pattern_line_colors):
        if color is None:
            continue
        fill_status = pattern_line_occupancy[pattern_line_index]
        base_index = cumulative_sizes[pattern_line_index - 1] if pattern_line_index > 0 else 0
        color_offset = int(color) * (pattern_line_index + 1)
        for status in range(fill_status):
            index = base_index + status + color_offset
            indices.append(index)

    return indices

def indices_to_pattern_lines(indices):
    cumulative_sizes = np.cumsum([i * NUM_TILE_COLORS for i in range(1, 6)])

    pattern_line_occupancy = [0] * 5
    pattern_line_colors = [None] * 5

    for index in indices:
        for pattern_line_index in range(5):
            base_index = cumulative_sizes[pattern_line_index - 1] if pattern_line_index > 0 else 0
            size = (pattern_line_index + 1) * NUM_TILE_COLORS

            if base_index <= index < base_index + size:
                relative_index = index - base_index
                color = relative_index // (pattern_line_index + 1)

                pattern_line_occupancy[pattern_line_index] += 1
                pattern_line_colors[pattern_line_index] = TileColor(color)
                break

    return pattern_line_occupancy, pattern_line_colors

def get_center_factory_indices(factories):
    center_factory = factories[-1]
    indices = []
    for color, count in enumerate(center_factory):
        indices.append(color * 10 + count)
    return indices

def center_factory_from_indices(indices):
    center_factory = np.zeros(NUM_TILE_COLORS, dtype=int)
    for index in indices:
        color = index // 10
        count = index % 10
        center_factory[color] = count
    return center_factory

def get_wall_indices(wall):
    indices = []
    for row in range(5):
        index = wall >> (6 * row) & 0b11111
        indices.append(index + 32 * row)

    for col in range(5):
        index = 0
        for row in range(5):
            bit = (wall >> (6 * row + col)) & 1
            index |= bit << row
        indices.append(index + 32 * col + 2 ** 5 * 5)

    for diag_index, mask in enumerate(WALL_COLOR_MASKS):
        index = 0
        masked = wall & mask
        for row in range(5):
            bit = (masked >> (6 * row)) & 0b11111
            index |= bit
        indices.append(index + 32 * diag_index + 32 * 10)

    return indices

def wall_from_indices(indices):
    wall = 0

    # Process rows
    wall1 = 0
    for row in range(5):
        index = indices[row] % 32
        wall |= index << (6 * row)
        wall1 |= index << (6 * row)

    # Process columns
    wall2 = 0
    for col in range(5):
        index = (indices[5 + col] - 32 * col - 2 ** 5 * 5) % 32
        for row in range(5):
            bit = (index >> row) & 1
            wall |= bit << (6 * row + col)
            wall2 |= bit << (6 * row + col)

    # Process diagonal indices
    wall3 = 0    
    for diag_index in range(len(WALL_COLOR_MASKS)):
        index = indices[10 + diag_index]
        base_index = 32 * diag_index + 32 * 10
        relative_index = index - base_index
        mask = WALL_COLOR_MASKS[diag_index]
        for row in range(5):
            for bit_index in range(5):
                bit_position = 6 * row + bit_index
                if mask & (1 << bit_position):
                    bit = (relative_index >> bit_index) & 1
                    wall |= bit << bit_position
                    wall3 |= bit << bit_position

    # Make sure all walls are equal. If not find the one that is different
    all_equal = wall == wall1 == wall2 == wall3
    if not all_equal:
        print(f"Wall : {wall:0>30b}")
        print(f"Wall1: {wall1:0>30b}")
        print(f"Wall2: {wall2:0>30b}")
        print(f"Wall3: {wall3:0>30b}")
        
        if wall1 != wall2:
            print("Wall1 != Wall2")
        if wall1 != wall3:
            print("Wall1 != Wall3")
        if wall2 != wall3:
            print("Wall2 != Wall3")

    return wall

def encode(game_state):
    encoding = np.zeros(TOTAL_ENCODING_SIZE, dtype=int)

    factory_indices = get_factory_indices(game_state.factories)
    for index in factory_indices:
        encoding[index] = 1

    center_factory_indices = get_center_factory_indices(game_state.factories)
    for index in center_factory_indices:
        encoding[FACTORY_ENCODING_SIZE + index] = 1
    
    for index in range(NUM_PLAYERS):
        player = (int(game_state.current_player) + index) % NUM_PLAYERS
        pattern_line_occupancy, pattern_line_colors = game_state.pattern_line_occupancy[player], game_state.pattern_line_colors[player]
        indices = pattern_lines_to_indices(pattern_line_occupancy, pattern_line_colors)
        for index in indices:
            encoding[FACTORY_ENCODING_SIZE + CENTER_FACTORY_ENCODING_SIZE + player * PATTERN_LINE_ENCODING_SIZE + index] = 1

        wall_indices = get_wall_indices(game_state.walls[player])
        for index in wall_indices:
            encoding[FACTORY_ENCODING_SIZE + CENTER_FACTORY_ENCODING_SIZE + player * PATTERN_LINE_ENCODING_SIZE + index] = 1
        
    return encoding

# def decode(encoding):
#     game_state = GameState()

#     # Decode factories
#     factory_indices = []
#     for index in range(FACTORY_ENCODING_SIZE):
#         if encoding[index] == 1:
#             factory_indices.append(index)
#     factories = factories_from_indices(factory_indices)

#     # Decode center factory
#     center_factory_indices = []
#     for index in range(CENTER_FACTORY_ENCODING_SIZE):
#         if encoding[FACTORY_ENCODING_SIZE + index] == 1:
#             center_factory_indices.append(index)
#     center_factory = center_factory_from_indices(center_factory_indices)
#     factories[-1] = center_factory
#     game_state.set_factories(factories)

#     # Decode pattern lines and walls for each player
#     pattern_line_occupancy = []
#     pattern_line_colors = []
#     walls = []
#     for player in range(NUM_PLAYERS):
#         player_offset = FACTORY_ENCODING_SIZE + CENTER_FACTORY_ENCODING_SIZE + player * (PATTERN_LINE_ENCODING_SIZE + WALL_ENCODING_SIZE)

#         # Decode pattern lines
#         pattern_line_indices = []
#         for index in range(PATTERN_LINE_ENCODING_SIZE):
#             if encoding[player_offset + index] == 1:
#                 pattern_line_indices.append(index)
#         player_pattern_line_occupancy, player_pattern_line_colors = indices_to_pattern_lines(pattern_line_indices)
#         # game_state.pattern_line_occupancy[player] = pattern_line_occupancy
#         # game_state.pattern_line_colors[player] = pattern_line_colors
#         pattern_line_occupancy.append(player_pattern_line_occupancy)
#         pattern_line_colors.append(player_pattern_line_colors)

#         # Decode walls
#         wall_indices = []
#         wall_offset = player_offset + PATTERN_LINE_ENCODING_SIZE
#         for index in range(WALL_ENCODING_SIZE):
#             if encoding[wall_offset + index] == 1:
#                 wall_indices.append(index)
#         wall = wall_from_indices(wall_indices)
#         walls.append(wall)
    
#     game_state.set_pattern_line_occupancy(pattern_line_occupancy)
#     game_state.set_pattern_line_colors(pattern_line_colors)
#     game_state.set_walls(walls)

#     return game_state

game_state = GameState()
for i in range(40):
    moves, _ = game_state.get_possible_moves()
    move = random.choice(moves)
    game_state.do_move(move)

# new = GameState()
# factories = factories_from_indices(get_factory_indices(game_state.factories))
# center = center_factory_from_indices(get_center_factory_indices(game_state.factories))
# factories[-1] = center
# new.set_factories(factories)

# pattern_line_occupancy = []
# pattern_line_colors = []
# walls = []
# for player in range(NUM_PLAYERS):
#     occupancy, colors = indices_to_pattern_lines(pattern_lines_to_indices(game_state.pattern_line_occupancy[player], game_state.pattern_line_colors[player]))
#     pattern_line_occupancy.append(occupancy)
#     pattern_line_colors.append(colors)
#     walls.append(wall_from_indices(get_wall_indices(game_state.walls[player])))

# new.set_pattern_line_occupancy(pattern_line_occupancy)
# new.set_pattern_line_colors(pattern_line_colors)
# new.set_walls(walls)


# print(game_state)
# print(new)
# assert game_state.walls == new.walls
# TOTAL_ENCODING_SIZE

encoding = encode(game_state)
encoding.shape
```

```python
# MOVE_ENCODING_SIZE = (len(INDEX_TO_FACTORY) + 1) * NUM_TILE_COLORS * 6 # 6 for pattern line 1-5 and discard, factory encoding + center and their colors

# def encode_move(game_state, move):
#     encoding = np.zeros(MOVE_ENCODING_SIZE, dtype=int)
#     factory_index = move.factory_index
#     tile_color = int(move.tile_color)
#     pattern_line_index = move.pattern_line_index
    
#     if factory_index < NUM_FACTORIES - 1:
#         factory = game_state.factories[factory_index]
#         factory_index = FACTORY_TO_INDEX[tuple(factory)]
#     else:
#         factory_index = len(INDEX_TO_FACTORY)
    
#     encoding[factory_index * NUM_TILE_COLORS + tile_color] = 1

move_index_set = set() # key is (factory_hash, tile_color, pattern_line_index)
for factory_index, factory in enumerate(INDEX_TO_FACTORY):
    for tile_color in range(NUM_TILE_COLORS):
        if factory[tile_color] == 0:
            continue

        for pattern_line_index in range(6):
            move_index_set.add((factory_index, tile_color, pattern_line_index))

for tile_color in range(NUM_TILE_COLORS):
    for pattern_line_index in range(6):
        move_index_set.add((len(INDEX_TO_FACTORY), tile_color, pattern_line_index))

MOVE_ENCODING_SIZE = len(move_index_set)
MOVE_TO_INDEX = {key: i for i, key in enumerate(move_index_set)}
INDEX_TO_MOVE = {i: key for key, i in MOVE_TO_INDEX.items()}

def encode_move(game_state, move):
    factory_index = move.factory_index
    tile_color = int(move.color)
    pattern_line_index = move.pattern_line_index

    if factory_index < NUM_FACTORIES - 1:
        factory = game_state.factories[factory_index]
        factory_index = FACTORY_TO_INDEX[tuple(factory)]
    else:
        factory_index = len(INDEX_TO_FACTORY)

    key = (factory_index, tile_color, pattern_line_index)
    move_index = MOVE_TO_INDEX.get(key)
    encoding = np.zeros(MOVE_ENCODING_SIZE, dtype=int)
    encoding[move_index] = 1
    return encoding

def decode_move(game_state, encoding):
    # Initialize a variable to store the best move and its evaluation score
    best_move = None
    best_score = -np.inf

    # Get the list of possible moves from the game state
    possible_moves = game_state.get_possible_moves()[0]

    # Iterate over all possible moves
    for move in possible_moves:
        # Extract the key for this move
        if move.factory_index < NUM_FACTORIES - 1:
            factory_key = tuple(game_state.factories[move.factory_index])
            factory_index = FACTORY_TO_INDEX[factory_key]
            key = (factory_index, int(move.color), move.pattern_line_index)
            move_index = MOVE_TO_INDEX[key]
        else:
            factory_index = len(INDEX_TO_FACTORY)
            key = (factory_index, int(move.color), move.pattern_line_index)
            move_index = MOVE_TO_INDEX[key]
        # Find the index of this move in the encoding
        # try:
        # Evaluate the score of this move
        score = encoding[move_index]
        # except KeyError:
        #     # If the key is not found in INDEX_TO_MOVE, continue to the next move
        #     continue
        
        # Update the best move if this move has a higher score
        if score > best_score:
            best_move = move
            best_score = score

    # If no valid move is found, return None or raise an exception
    if best_move is None:
        raise ValueError("No valid move found based on the given encoding.")
    
    return best_move
```

```python
import torch
import torch.nn as nn

class AzulNnue(nn.Module):
    def __init__(self):
        super(AzulNnue, self).__init__()
        self.layers = nn.Sequential(
            nn.Linear(TOTAL_ENCODING_SIZE, 2048),
            nn.ReLU(),
            nn.Linear(2048, 256),
            nn.ReLU(),
            # nn.Linear(2048, 2048),
            # nn.ReLU(),
            # nn.Linear(2048, 2048),
            # nn.ReLU(),
            # nn.Linear(2048, 2048),
            # nn.ReLU(),
            # nn.Linear(2048, 2048),
            # nn.ReLU(),
            # nn.Linear(2048, 1024),
            # nn.ReLU(),
            # nn.Linear(1024, 512),
            # nn.ReLU(),
            # nn.Linear(512, 256),
            # nn.ReLU(),
            nn.Linear(256, MOVE_ENCODING_SIZE),
        )

    def forward(self, x):
        x = self.layers(x)
        return x

    def get_move(self, game_state):
        encoding = encode(game_state)
        tensor_encoding = torch.tensor(encoding, dtype=torch.float32).clone().detach()
        output = self(tensor_encoding).clone().detach()
        return decode_move(game_state, output)

model = AzulNnue()
loss_history = []
# game_state = GameState()
# print(game_state)

# move = model.get_move(game_state)
# print(move)

# game_state.do_move(move)
# print(game_state)
```

```python
states = []
actions = []
for state, action in state_action_pairs:
    state_encoding = encode(state)
    action_encoding = encode_move(state, action)
    states.append(state_encoding)
    actions.append(action_encoding)
```

```python
from torch.utils.data import DataLoader, TensorDataset

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)

states_tensor = torch.tensor(states, dtype=torch.float32).to(device)
actions_tensor = torch.tensor(actions, dtype=torch.float32).to(device)

dataset = TensorDataset(states_tensor, actions_tensor)
dataloader = DataLoader(dataset, batch_size=128, shuffle=True)
```

```python
import torch.optim as optim
import matplotlib.pyplot as plt
from IPython.display import clear_output

model.to(device)

# Define the loss function and optimizer
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Training loop
num_epochs = 160
counter = 0

try:
    for epoch in range(num_epochs):
        for batch_states, batch_actions in dataloader:
            optimizer.zero_grad()
            outputs = model(batch_states)
            loss = criterion(outputs, batch_actions)
            loss.backward()
            optimizer.step()

            counter += 1
            if counter % 10 == 0:
                print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item()}" + " " * 10, end="\r")
                loss_history.append(loss.item())

        print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item()}")
except KeyboardInterrupt:
    pass
model.to("cpu")
```
