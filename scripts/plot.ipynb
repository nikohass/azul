```python
# Importing required libraries
import json

# File path
file_path = '../logs/automated_test.log'

# Reading and processing data
think_times = []
scores_player1 = []
scores_player2 = []

# Open and read the file
with open(file_path, 'r') as file:
    for line in file:
        data = json.loads(line)  # Load each line as a JSON object
        config = data['game_config']
        result = data['result']
        
        # Extracting data for each player
        player1_time = config['players'][0]['think_time']
        player2_time = config['players'][1]['think_time']
        player1_score = result['scores'][0]
        player2_score = result['scores'][1]
        
        # Appending to lists
        if player1_time == 1000:  # Only add when Player 1's time is 1000ms
            think_times.append(player2_time)
            scores_player1.append(player1_score)
            scores_player2.append(player2_score)

# Plotting the scores over think times
plt.figure(figsize=(10, 5))
plt.plot(think_times, scores_player1, label='Spieler 1', marker='o')
plt.plot(think_times, scores_player2, label='Spieler 2', marker='o')
plt.title('Punkteverhältnis über die Zeit')
plt.xlabel('Denkzeit in ms')
plt.ylabel('Durchschnittliche Punktzahl')
plt.grid(True)
plt.legend()
plt.show()
```

```python
think_times = []
win_ratios_player1 = []

# Open and read the file
with open(file_path, 'r') as file:
    for line in file:
        data = json.loads(line)  # Load each line as a JSON object
        config = data['game_config']
        result = data['result']
        
        # Extracting data for each player
        player1_time = config['players'][0]['think_time']
        player2_time = config['players'][1]['think_time']
        player1_wins = result['wins'][0]
        player2_wins = result['wins'][1]
        
        # Appending to lists
        if player1_time == 1000:  # Only add when Player 1's time is 1000ms
            think_times.append(player2_time)
            # Calculate win ratio of player 1 to player 2
            win_ratio = player1_wins / (player1_wins + player2_wins) if (player1_wins + player2_wins) > 0 else 0
            win_ratios_player1.append(win_ratio)

# Plotting the win ratios over think times
plt.figure(figsize=(10, 5))
plt.plot(think_times, win_ratios_player1, label='Siegverhältnis Spieler 1', marker='o', color='green')
plt.title('Siegverhältnis von Spieler 1 über die Zeit')
plt.xlabel('Denkzeit in ms')
plt.ylabel('Siegverhältnis Spieler 1')
plt.grid(True)
plt.legend()
plt.show()
```

```python
think_times = []
score_ratios = []
with open(file_path, 'r') as file:
    for line in file:
        data = json.loads(line)
        config = data['game_config']
        result = data['result']
        
        player1_time = config['players'][0]['think_time']
        player2_time = config['players'][1]['think_time']
        player1_score = result['scores'][0]
        player2_score = result['scores'][1]
        
        # Daten nur hinzufügen, wenn die Denkzeit von Spieler 1 1000ms beträgt
        if player1_time == 1000:
            think_times.append(player2_time)
            # Punkteverhältnis von Spieler 1 zu Spieler 2 berechnen
            if player2_score > 0:  # Um Division durch Null zu vermeiden
                score_ratio = player1_score / player2_score
            else:
                score_ratio = 0
            score_ratios.append(score_ratio)

# Erstellen des Plots für das Punkteverhältnis
plt.figure(figsize=(10, 5))
plt.plot(think_times, score_ratios, label='Punkteverhältnis Spieler 1 zu Spieler 2', marker='o', color='blue')
plt.title('Punkteverhältnis von Spieler 1 zu Spieler 2 über die Zeit')
plt.xlabel('Denkzeit in ms')
plt.ylabel('Punkteverhältnis')
plt.grid(True)
plt.legend()
plt.show()
```
