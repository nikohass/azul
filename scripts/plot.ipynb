```python
import re
import networkx as nx
import matplotlib.pyplot as plt

def parse_node_line(line):
    """ Parses a node line and returns node id and attributes. """
    node_pattern = r"Node (\d+): \{ event: (\w+)\((.*?)\), n: ([\d.]+), q: (.*?), game_over: (.*?), prob_children: (.*?) \}"
    match = re.match(node_pattern, line)
    if match:
        node_id = int(match.group(1))
        event_type = match.group(2)
        event_details = match.group(3)
        n = float(match.group(4))
        q_values = match.group(5).split()  # Split the q string into a list of strings
        q_floats = [float(q) for q in q_values]  # Convert q strings to floats
        if n != 0:  # To avoid division by zero
            q_normalized = [round(q / n, 2) for q in q_floats]  # Divide each q by n
        else:
            q_normalized = q_floats  # If n is 0, use the original q values to avoid division by zero
        game_over = match.group(6) == 'true'
        prob_children = match.group(7) == 'true'
        return node_id, {
            'event_type': event_type,
            'event_details': event_details,
            'n': n,
            'q': q_normalized,
            'game_over': game_over,
            'prob_children': prob_children
        }
    return None, None

def parse_edge_line(line):
    """ Parses an edge line and returns the tuple of connected nodes. """
    edge_pattern = r"Edge from Node (\d+) to Node (\d+)"
    match = re.match(edge_pattern, line)
    if match:
        return int(match.group(1)), int(match.group(2))
    return None

def read_data(file_path):
    """ Reads the tree data from file and returns nodes and edges. """
    nodes = {}
    edges = []
    with open(file_path, 'r') as file:
        for line in file:
            line = line.strip()
            if line.startswith('Node'):
                node_id, node_data = parse_node_line(line)
                if node_data:
                    nodes[node_id] = node_data
            elif line.startswith('Edge'):
                edge = parse_edge_line(line)
                if edge:
                    edges.append(edge)
    return nodes, edges

def build_graph(nodes, edges):
    """ Constructs a directed graph from nodes and edges data. """
    G = nx.DiGraph()
    for node_id, attributes in nodes.items():
        G.add_node(node_id, **attributes)
    for edge in edges:
        G.add_edge(*edge)
    return G

file_path = "../logs/tree.txt"
nodes, edges = read_data(file_path)
G = build_graph(nodes, edges)
print(f"Number of nodes: {G.number_of_nodes()}")
print(f"Number of edges: {G.number_of_edges()}")
```

```python
CURRENT_PLAYER = 1

def visualize_graph(G):
    """Visualizes the graph using networkx and matplotlib with a hierarchical tree layout if it's a DAG."""
    if nx.is_directed_acyclic_graph(G):
        # Generate position information using a hierarchical layout for DAGs
        pos = nx.nx_agraph.graphviz_layout(G, prog='sfdp')
    else:
        # Fallback to spring layout if not a DAG
        pos = nx.spring_layout(G)

    labels = {node: f"{data['q'][CURRENT_PLAYER]}" for node, data in G.nodes(data=True)}
    # labels = {node: f"q={data['q'][CURRENT_PLAYER]}\nn={int(data['n'])}" for node, data in G.nodes(data=True)}
    # node_colors = ['red' if data['game_over'] else 'salmon' for node, data in G.nodes(data=True)]
    # node_colors = ['skyblue' if data['event_type'] == "Probabilistic" else 'salmon' for node, data in G.nodes(data=True)]
    node_colors = []
    is_root = True
    for node, data in G.nodes(data=True):
        if is_root:
            node_colors.append('blue')
            is_root = False
        elif data['game_over']:
            node_colors.append('red')
        elif data['event_type'] == "Probabilistic":
            node_colors.append('orange')
        else:
            node_colors.append('green')
    plt.figure(figsize=(40, 40))
    nx.draw(G, pos, with_labels=False, node_color=node_colors, alpha=0.4, arrows=True, node_size=50, edge_color='black', font_size=1)
    nx.draw_networkx_labels(G, pos, labels=labels, font_size=8)
    plt.show()

visualize_graph(G)
```
